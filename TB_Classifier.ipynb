# **Tuberculosis Classifier**

I have build a machine learning model that classifies TB using images. I have used this data [https://www.kaggle.com/tawsifurrahman/tuberculosis-tb-chest-xray-dataset](https://www.kaggle.com/tawsifurrahman/tuberculosis-tb-chest-xray-dataset)


## **Loading Necessary Libraries.**

```python
import numpy  as np
import pandas as pd
from skimage.transform import resize
from skimage.io import imread
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

## Converting images into numpy array and then loading it into a dataframe

```python
Root_dir='/content/drive/MyDrive/TB_Chest_Radiography_Database'
```

```python
Categories=['Normal', 'Tuberculosis']
feature_arr=[]
labels_arr=[]
```

```python
for i in Categories:
  path=os.path.join(Root_dir, i)
  for img in os.listdir(path):
    img_array=imread(os.path.join(path, img))
    img_resize=resize(img_array,(150,150,3))
    feature_arr.append(img_resize.flatten())
    labels_arr.append(Categories.index(i))
  print('Loading done')
```

```python
flat_features=np.array(feature_arr)
labels=np.array(labels_arr)
```

```python
df = pd.DataFrame(flat_features)
df["Label"] = labels
```

```python
df.head()
```

```python
final_features=df.drop(['Label'], axis='columns')
```

```python
final_features.head()
```

```python
final_labels=df[['Label']]
```

```python
final_labels.head()
```

## Building Machine Learning Models

I have used 3 machine learning models.
1. KNN had an accuracy of 94%.
2. Decision Trees had an accuracy of 94%.
3. RandomForest Classifer had the best accuracy of 98%.


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test=train_test_split(final_features, final_labels, train_size=0.7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
KNN=KNeighborsClassifier()
```

```python
KNN_model=KNN.fit(x_train, y_train)
```

```python
print(f" KNN  ACCURACY {KNN_model.score(x_test,y_test)}")
```

```python
KNN_predic=KNN_model.predict(x_test)
```

```python
Tree = DecisionTreeClassifier()
Tree.fit(x_train,y_train)
```

```python
print(f" Decision Tree accuracy {Tree.score(x_test,y_test)}")
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
print(f"Random Forest accuracy : {rf.score(x_test,y_test)}")
```

```python
best_predict=rf.predict(x_test)
```
